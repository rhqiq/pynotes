```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') # More on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.0, 0.0, 1., 1.]) # main axes
axes2 = fig.add_axes([0.5, 0.5, 0.4, 0.4]) # inset axes
axes3 = fig.add_axes([0.1, 0.1, 0.3, 0.3]) # inset axes
```

```python
fig, axes = plt.subplots(2,2)
axes[0][0].plot(x,y)
axes[1][1].plot(y,x)
```

```python
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

# Figure size, aspect ratio and DPI

```python
fig, axes = plt.subplots(figsize=(12,4))

axes.plot(x, y, 'r')
axes.plot(y, x, 'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
plt.tight_layout()
```

```python
fig.savefig("filename.png", dpi=200)
```
